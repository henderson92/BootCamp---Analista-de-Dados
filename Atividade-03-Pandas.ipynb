```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Base de dados - Dados sobre desgaste de funcionarios.csv")
df
```

### Distribuição etária : Extraia a distribuição de idades usando a Idade coluna.

```python
# Filtra por idade acima de 40 anos
df_filtra_idade = df[df['Idade'] > 40]
print(df_filtra_idade)
```

### Nível de escolaridade : Analisar a distribuição dos Escolaridadeníveis.

```python
df_escolaridade = df[df['Escolaridade'] == 2]
print(df_escolaridade)
```

```python
print(df['Escolaridade'].value_counts())
```

```python
print(df['Departamento'].value_counts())
```

```python
print("Total de anos de experiência de trabalho:")
print(df['Total de anos trabalhados'].describe())
print("Anos trabalhados na empresa atual:")
print(df['Total de anos trabalhados na empresa'].describe())
print("Anos trabalhados na função atual:")
print(df['Anos trabalhados na função atual'].describe())
print("Anos desde a última promoção:")
print(df['Anos desde a última promoção'].describe())
```

```python
print("Satisfação com o equilíbrio entre vida pessoal e profissional:")
print(df['Equilíbrio vida trabalho'].describe())
print("Satisfação com relacionamentos:")
print(df['Satisfação nas relações Não trabalho'].describe())
print("Satisfação geral no trabalho:")
print((df['Equilíbrio vida trabalho'] + df['Satisfação nas relações Não trabalho']).describe())
```

```python
print("Remuneração variável:")
print(df['opção de remaunração variável em ações'].value_counts())
print("Salário médio:")
print(df['Diárias recebidas'].mean())
```

```python
print("Distância de casa:")
print(df['Distância de Casa'].describe())
print("Frequência de viagens de negócios:")
print(df['Viagem de negócios'].value_counts())
```

```python
print("Número de sessões de treinamento:")
print(df['Qtde de treinamentos realizados no último ano'].describe())
print("Frequência de treinamento:")
print(df['Qtde de treinamentos realizados no último ano'].mean())
```

```python
print("Retenção de funcionários:")
print(df['Número de contratos de trabalho na empresa'].describe())
print("Tempo médio de serviço:")
print(df['Total de anos trabalhados na empresa'].mean())
```

```python
print("Anos trabalhados com o gerente atual:")
print(df['Anos trabalhando com o atual gestor'].describe())
print("Experiência gerencial:")
print(df.groupby('Anos trabalhados na função atual')['Anos trabalhando com o atual gestor'].describe())
```

```python
# Análise de Gênero
print("Distribuição de gênero:")
print(df['Gênero'].value_counts())
```

```python
# Análise de Idade x Gênero
plt.hist(df[df['Gênero'] == 'Masculino']['Idade'], bins=10, alpha=0.5, label='Masculino')
plt.hist(df[df['Gênero'] == 'Feminino']['Idade'], bins=10, alpha=0.5, label='Feminino')
plt.xlabel('Idade')
plt.ylabel('Frequency')
plt.title('Distribuição de Idade por Gênero')
plt.legend()
plt.show()
```

```python
# Análise de Departamento x Gênero
print("Distribuição de departamentos por gênero:")
print(pd.crosstab(df['Departamento'], df['Gênero']))
```

```python
# Análise de Experiência x Gênero
print("Distribuição de experiência por gênero:")
print(pd.crosstab(df['Total de anos trabalhados'], df['Gênero']))
```

```python
# Análise de Treinamento x Gênero
print("Distribuição de treinamento por gênero:")
print(pd.crosstab(df['Qtde de treinamentos realizados no último ano'], df['Gênero']))
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x="Departamento", y="Idade", data=df)
plt.title('Distribuição de Idade por Departamento')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(x="Departamento", y="Idade", data=df)
plt.title('Distribuição de Idade por Departamento')
plt.show()
```

```python
# Análise de Jornada padrão de trabalho x Área de Formação
plt.figure(figsize=(10,6))
sns.countplot(x="Jornada padrão de trabalho", hue="Área de Formação", data=df)
plt.title('Distribuição de Área de Formação por Jornada Padrão de Trabalho')
plt.show()
```

```python
# Análise de Demissão x Departamento
plt.figure(figsize=(10,6))
sns.countplot(x="Departamento", hue="Demissão", data=df)
plt.title('Distribuição de Demissão por Departamento')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x="Demissão", data=df)
plt.title('Distribuição de Demissão')
plt.xlabel('Demissão')
plt.ylabel('Contagem')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="Demissão", data=df, estimator=lambda x: sum(x)/len(x))
plt.title('Proporção de Demissão')
plt.xlabel('Demissão')
plt.ylabel('Proporção')
plt.show()
```
